# QLED and OLED TVs from jumia

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import urllib
# my_url="https://www.jumia.com.ng/oled-tvs/"
# uClient = uReq(my_url)
# page_html = uClient.read()
# uClient.close()
```

```python
my_url = "https://www.jumia.com.ng/oled-tvs/"
header= {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' 
      'AppleWebKit/537.11 (KHTML, like Gecko) '
      'Chrome/23.0.1271.64 Safari/537.11',
      'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
      'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
      'Accept-Encoding': 'none',
      'Accept-Language': 'en-US,en;q=0.8',
      'Connection': 'keep-alive'}

#the URL where you are requesting at
req = urllib.request.Request(url=my_url, headers=header) 
page_html = urllib.request.urlopen(req).read()

uClient = uReq(req)
page_html = uClient.read()
uClient.close()

page_soup = soup(page_html, "html.parser")
page_soup
```

```python
containers = page_soup.findAll("div", { "id": "jm"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
container = containers[0]
contents = containers[0].findAll("a", {"class": "core"})
len(contents)
```

```python
prod_cont = contents[0]
prod_cont
```

```python
# Scrape product image
product_image = prod_cont.findAll("img", {"class": "img"})
print(product_image[0]['src'])

# Scrape product name
product_name = prod_cont.findAll("h3", {"class": "name"})
print(product_name[0].text)

# product price
price = prod_cont.findAll("div", {"class": "prc"})
print(price[0].text)

# old price
old_price = prod_cont.findAll("div", {"class": "old"})
print(old_price[0].text)

# discount
discount = prod_cont.findAll("div", {"class": "bdg _dsct _sm"})
print(discount[0].text)
```

```python
products_image =[]
products_name = []
pricing = []
old_price = []
discount = []
```

```python
for content in contents:
    image = content.findAll("img", {"class": "img"})
    products_image.append(image[0]['src'])
    
    name = content.findAll("h3", {"class": "name"})
    products_name.append(name[0].text)
    
    price = prod_cont.findAll("div", {"class": "prc"})
    pricing.append(price[0].text)
    
    old = prod_cont.findAll("div", {"class": "old"})
    if len(old) == 0:
        old_price.append(np.nan)
    else:
        old_price.append(old[0].text)
    
    disc = prod_cont.findAll("div", {"class": "bdg _dsct _sm"})
    if len (disc) == 0:
        discount.append(np.nan)
    else:
        discount.append(disc[0].text)

print(len(products_image))
print(len(products_name))
print(len(pricing))
print(len(old_price))
print(len(discount))
```

```python
# to scrape other pages of the product in the website
my_url = ['https://www.jumia.com.ng/oled-tvs/',
          'https://www.jumia.com.ng/oled-tvs/?page=2#catalog-listing',
          'https://www.jumia.com.ng/oled-tvs/?page=3#catalog-listing'
         ]
for links in my_url:
    
    req = urllib.request.Request(url=links, headers=header) 
    page_html = urllib.request.urlopen(req).read()
    
    page_soup = soup(page_html, 'html.parser')
    page_soup
    
    containers = page_soup.findAll("div", { "id": "jm"})
    
    container = containers[0]
    contents = containers[0].findAll("a", {"class": "core"})
    
    for content in contents:
        image = content.findAll("img", {"class": "img"})
        products_image.append(image[0]['src'])

        name = content.findAll("h3", {"class": "name"})
        products_name.append(name[0].text)

        price = prod_cont.findAll("div", {"class": "prc"})
        pricing.append(price[0].text)

        old = prod_cont.findAll("div", {"class": "old"})
        if len(old) == 0:
            old_price.append(np.nan)
        else:
            old_price.append(old[0].text)

        disc = prod_cont.findAll("div", {"class": "bdg _dsct _sm"})
        if len (disc) == 0:
            discount.append(np.nan)
        else:
            discount.append(disc[0].text)
            
print(len(products_image))
print(len(products_name))
print(len(pricing))
print(len(old_price))
print(len(discount))
```

```python
# save and view the data
df = pd.DataFrame({"Product Name":products_name, "Price":pricing, "Old Price":old_price, "Discount":discount, 
                   "Image":products_image})
df.to_csv("QLED and OLED TVs.csv")
tvs = pd.read_csv("QLED and OLED TVs.csv")

print(tvs.shape)

tvs.head()
```
